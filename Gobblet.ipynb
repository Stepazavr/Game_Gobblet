FIELD

```python
class Field(ABC):
    def __init__(self, Point, Size, colour, Data):
        self.Point = Point
        self.Size = Size
        self.colour = colour
        self.Data = Data
        
        self.Blocks = np.empty((self.Data.number, self.Data.number), dtype=Block)      

#Engine
    @abstractmethod
    def start_init(self):
        self.init_all_blocks()   

#Engine
    def init_point_to_block(self, row, col):
        x = self.Point.x + col * Set.block_side.value + (col + 1) * Set.block_margin.value
        y = self.Point.y + row * Set.block_side.value + (row + 1) * Set.block_margin.value 
        return Point(x, y)

#Engine
    def init_block(self, row, col, Point, Size, colour):
        self.Blocks[row][col] = Block(Point, Size, colour)                            
    
#Engine
    def init_all_blocks(self):
        for row in range(self.Data.number):
            for col in range(self.Data.number):
                block_point = self.init_point_to_block(row, col) 
                block_colour = self.init_colour_to_block(row + col)
                block_Size = Size(Set.block_side.value, Set.block_side.value)
                self.init_block(row, col, block_point, block_Size, block_colour)
    
#Engine 
    @abstractmethod
    def mouse_update(self):
        pass
    
#Engine 
    @abstractmethod
    def init_colour_to_block(self, number):
        pass

#------------------------------------------
#Engine
    @abstractmethod
    def check_draw(self):
        pass
    
#Engine
    @abstractmethod
    def check_rows(self):
        pass
    
#Engine
    @abstractmethod
    def check_columns(self):
        pass
            
#Engine
    @abstractmethod
    def check_diagonals_left(self):
        pass
    
#Engine
    @abstractmethod
    def check_diagonals_right(self):
        pass

#Engine
    def update_score(self):
        for i in range(len(self.Data.Players)):
            if self.Data.Players[i].win:
                self.Data.score[i] += 1

#Engine
    def update_main_queue(self):
        self.Data.main_queue ^= 1

#Engine
    def update_queue(self):
        self.Data.queue ^= 1
                
#Engine
    def get_text_to_game_over(self):
        for Player in self.Data.Players:
            if Player.win:
                return self.Data.Fonts[1].render(Const.Win_phase[0] +  Player.name + Const.Win_phase[1], True, Const.FINISH_COLOUR)
        return self.Data.Fonts[1].render(Const.Draw_phase, True, Const.FINISH_COLOUR)           
    

#Engine
    def set_Player_win(self, win_colour):
        for Player in self.Data.Players:
            if Player.colour == win_colour:
                Player.win = True     

#Engine        
    def check_index(self, row, col):
        if -1 < row < self.Data.number and  -1 < col < self.Data.number:
            return True
        return False  


#Engine
    def win_check(self):
         return self.check_rows() or self.check_columns() or self.check_diagonals_left() or self.check_diagonals_right()

#------------------------------------------ 

        
```

FIELD_GOBBLET

```python
class Field_Gobblet(Field):
    def __init__(self, Point, Size, colour, Info):
        super().__init__(Point, Size, colour, Info)
        
#Engine      
    def start_init(self):
        super().start_init()
        self.init_counter_points()
        
    
#Engine
    def add_point_to_counter(self, counter, point):
        counter.Point.x = point.x
        counter.Point.y = point.y

#Engine 
    def init_counter_points(self):
        for Player in self.Data.Players:
            for j in range(len(Player.Counters)):
                if Player.Counters[j].colour == Const.COUNTER_COLOURS[0]:
                    center_x = self.Point.x // 2
                else:
                    center_x = Set.window_w.value - self.Point.x // 2                
                center_y = Set.window_h.value // (Set.big_radius.value + 2) * (j // 3 + 2)
                self.add_point_to_counter(Player.Counters[j], Point(center_x, center_y))
                
    def init_colour_to_block(self, number):
        return Const.BLOCK_COLOURS[number % 2]
#-------------------------------------------                

#Engine
    def check_draw(self):
        pass    

    def check_lines(self, Blocks):
        for row in range(self.Data.number):
            if self.block_empty(Blocks[row][0]):
                continue
            current = True
            base_colour = Blocks[row][0].Counter.colour
            for col in range(1, self.Data.number):
                if self.block_empty(Blocks[row][col]):
                    current = False
                    break
                elif Blocks[row][col].Counter.colour != base_colour:
                    current = False
                    break
            if current:
                self.set_Player_win(base_colour)
                return True
        return False
 


    def check_rows(self):
        return self.check_lines(self.Blocks)
    
    def check_columns(self):
        return self.check_lines(self.Blocks.copy().transpose())

    def check_diagonals(self, Blocks):
        if self.block_empty(Blocks[0][0]):
            return False
        base_colour = Blocks[0][0].Counter.colour
        for i in range(1, self.Data.number):
            if self.block_empty(Blocks[i][i]):
                return False
            if Blocks[i][i].Counter.colour != base_colour:
                return False
        self.set_Player_win(base_colour)
        return True   

    def check_diagonals_left(self):
        return self.check_diagonals(self.Blocks)

    def check_diagonals_right(self):
        return self.check_diagonals(np.fliplr(self.Blocks.copy()))
        
            
#Engine  
    def is_counter_on_field(self, counter):
        return (counter.Point.x > self.Point.x and counter.Point.x < self.Point.x + self.Size.width and counter.Point.y > self.Point.y and counter.Point.y < self.Point.y + self.Size.height)

    def get_block_with_counter(self, counter):
        for row in self.Blocks:
            for block in row:
                if block.Counter is counter:
                    return block
    
    def break_the_conection_of_block_and_counter(self, block):
        block.Counter = None
        
    def close_block(self, block):
        block.close = True
        
    def delete_counter(self, block):
        for Player in self.Data.Players:
            for counter in Player.Counters:
                if counter is block.Counter:
                    block.Counter = None
                    Player.Counters.remove(counter)
                    return True
                         

#Engine
    def give_new_point_to_counter(self, counter, point):
        counter.Point.x = point.x + Set.block_side.value // 2
        counter.Point.y = point.y + Set.block_side.value // 2
        
#Engine
    def give_counter_to_block(self, block, counter):
        if self.is_counter_on_field(counter):
            old_block = self.get_block_with_counter(counter)
            self.break_the_conection_of_block_and_counter(old_block)
        else:
            self.reduce_numbers(counter)
        
        block.Counter = counter
        self.give_new_point_to_counter(block.Counter, block.Point)
                
    def reduce_numbers(self, counter):
        for p in range(len(self.Data.Players)):
            for curr in self.Data.Players[p].Counters:
                if curr is counter:
                    self.Data.Numbers[p][4 - curr.value] -= 1
                    return True
                
    def block_indexes(self, point):
        row =  int((point.y - self.Point.y) // (Set.block_side.value + Set.block_margin.value))
        col =  int((point.x - self.Point.x) //  (Set.block_side.value + Set.block_margin.value))        
        return Point(row, col)

    def check_counter(self, counter, block):
        if self.is_counter_on_field(counter):          
            old_block = self.get_block_with_counter(counter)
            if old_block and old_block.close or old_block is block:
                return False
        return True
        
    def update_block_and_counter(self, block, counter):
        if not self.block_empty(block) and block.Counter.value < counter.value and block.Counter.colour != counter.colour and block.Counter is not counter:
            self.close_block(block)
            self.delete_counter(block)
            self.give_counter_to_block(block, counter)
            self.update_queue()           
        elif self.block_empty(block):
            self.give_counter_to_block(block, counter)
            self.update_queue() 
        
            
#Engine                
    def update_block_mouse(self, counter, point):
        Indexes = self.block_indexes(point)
        row, col = Indexes.x, Indexes.y
        if not self.check_index(row, col):
            return False
        block = self.Blocks[row][col]

        if not self.check_counter(counter, block):
            return False
                    
        self.update_block_and_counter(block, counter)

#Engine       
    def block_empty(self, block):
        return (block.Counter == None)
            
#------------------------------------------
#Engine
    def mouse_update(self, counter):
        x_mouse, y_mouse = pygame.mouse.get_pos()
        self.update_block_mouse(counter, Point(x_mouse, y_mouse))
```

FIELD_CROSS

```python
class Field_Cross(Field):
    def __init__(self, Point, Size, colour, Info):
        super().__init__(Point, Size, colour, Info)                                   
                                      
    def start_init(self):
        super().start_init()
        
        
    def init_colour_to_block(self, number):     
        return Const.BLOCK_COLOUR
                
        
    def check_draw(self):
        for row in range(self.Data.number):
            for col in range(self.Data.number):
                if self.Blocks[row][col].colour == Const.BLOCK_COLOUR:
                    return False
        return True
            
                
    def check_lines(self, Blocks):
        for row in range(self.Data.number):
            base_colour = Blocks[row][0].colour
            current = True
            for col in range(1, self.Data.number):
                if Blocks[row][col].colour == Const.BLOCK_COLOUR or Blocks[row][col].colour != base_colour:
                    current = False
                    break
            if current:
                self.set_Player_win(base_colour)
                return True
        return False
    
    
    def check_rows(self):
        return self.check_lines(self.Blocks)
    
    def check_columns(self):
        return self.check_lines(self.Blocks.copy().transpose())

    
    def check_diagonals(self, Blocks):
        base_colour = Blocks[0][0].colour
        for i in range(1, self.Data.number):
            if Blocks[i][i].colour == Const.BLOCK_COLOUR or Blocks[i][i].colour != base_colour:
                return False
        self.set_Player_win(base_colour)
        return True

    
    def check_diagonals_left(self):
        return self.check_diagonals(self.Blocks)

    def check_diagonals_right(self):
        return self.check_diagonals(np.fliplr(self.Blocks.copy()))
    
    
    def block_empty(self, block):
        return (block.colour == Const.BLOCK_COLOUR)
                

    def update_block_mouse(self, point):
        col =  int((point.x - self.Point.x) //  (Set.block_side.value + Set.block_margin.value))
        row =  int((point.y - self.Point.y) // (Set.block_side.value + Set.block_margin.value))
        if not self.check_index(row, col):
            return False
        block = self.Blocks[row][col]       
        
        if self.block_empty(block):
            block.colour = self.Data.Players[self.Data.queue].field_colour
            self.update_queue()
        return True

    def mouse_update(self):
        x_mouse, y_mouse = pygame.mouse.get_pos()
        self.update_block_mouse(Point(x_mouse, y_mouse))
                    
```

GAME

```python
class Game(ABC):  
    def __init__(self, Window):        
        self.Data = None        
        
    def check_field(self):
        self.Data.win_game = self.Data.Field.win_check()
        self.Data.draw_game = self.Data.Field.check_draw()
        self.Data.game_over = self.Data.win_game or self.Data.draw_game
        if self.Data.game_over:
            self.Data.text_to_game_over = self.Data.Field.get_text_to_game_over()
        
    def event_cycle(self):
        self.check_field()
        for event in pygame.event.get():
            self.check_events(event) 

    def check_events(self, event):        
        if event.type == pygame.QUIT:
            self.Data.running = False 
            self.Data.main_running = False
        elif event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
            self.Data.return_game = True
            self.Data.running = False
        elif event.type == pygame.MOUSEBUTTONDOWN and not self.Data.game_over:
            x_mouse, y_mouse = pygame.mouse.get_pos()
            if 0 < x_mouse - Formula.Name_Point_x(Const.Name_Menu, Set.icon_menu_x.value) < Formula.Name_len(Const.Name_Menu) and 0 < y_mouse  - Set.icon_menu_y.value < Set.letter_height.value:
                self.Data.running = False
            else:
                self.mouse_button_down(event) 
            
    @abstractmethod
    def mouse_button_down(self, event):
        pass

    @abstractmethod
    def mouse_update(self):
        pass

    @abstractmethod
    def return_Data(self):
        pass
    
    def return_score(self):
        self.Data.score[0], self.Data.score[1] = self.Data.score[1], self.Data.score[0]

    def return_start_game(self):
        self.Data.Field.update_score()
        self.Data.Field.update_main_queue()
        self.return_Data()
        self.return_score()
```

```python
class Cross(Game): 
    def __init__(self, Window):
        self.Data = Data_cross(Window, 0, [0,0])
        
    def return_Data(self):
        self.Data = Data_cross(self.Data.Window, self.Data.main_queue, self.Data.score)         
            
    def mouse_button_down(self, event):
        self.Data.click_mouse = True
                             
    def mouse_update(self):
        if self.Data.click_mouse:
            self.Data.Field.mouse_update() 
        self.Data.click_mouse = False
        return True
```

GOBBLET

```python
class Gobblet(Game):
    def __init__(self, Window):
        self.Data = Data_gobblet(Window, 0, [0, 0])
        
    def return_Data(self):
        self.Data = Data_gobblet(self.Data.Window, self.Data.main_queue, self.Data.score)
        
    def mouse_point_to_counter(self, counter, point):
        if counter.Point.x - counter.Size.width < point.x < counter.Point.x + counter.Size.width:
            if counter.Point.y - counter.Size.height < point.y < counter.Point.y + counter.Size.height:  
                return True
        return False        
        
    def mouse_button_down(self, event):
        for counter in self.Data.Players[self.Data.queue].Counters:
            if self.mouse_point_to_counter(counter, Point(event.pos[0], event.pos[1])):
                self.Data.click_counter = counter
                self.Data.click_mouse = True
                return True
        self.Data.click_counter = None
        self.Data.click_mouse = False
        return True
                   
    def mouse_update(self):
        if not self.Data.click_mouse:
            return True
        flag = True
        while flag:
            for event in pygame.event.get():
                if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1 and not self.Data.game_over:
                    flag = False
                    break
        self.Data.Field.mouse_update(self.Data.click_counter)
        self.Data.click_counter = None
        self.Data.click_mouse = False
        return True
```

MENU

```python
class Menu:
    def __init__(self):
        self.Name_icon = False
        self.Ezic = False
        
    def check_menu_event(self):
        for event in pygame.event.get():
            if event.type == pygame.MOUSEBUTTONDOWN:
                x_mouse, y_mouse = pygame.mouse.get_pos()
                if 0 < x_mouse - Formula.Name_Point_x(Const.Name_Gobblet, Set.icon_Gobblet_x.value) < Formula.Name_len(Const.Name_Gobblet) and 0 < y_mouse  - Set.icon_Gobblet_y.value < Set.letter_height.value:
                    self.Name_icon = Const.Name_Gobblet
                    return True
                if 0 < x_mouse - Formula.Name_Point_x(Const.Name_Cross, Set.icon_Cross_x.value) < Formula.Name_len(Const.Name_Cross) and 0 < y_mouse  - Set.icon_Cross_y.value < Set.letter_height.value:
                    self.Name_icon = Const.Name_Cross
                    return True
                if 0 < x_mouse - Formula.Name_Point_x(Const.Name_Quit, Set.icon_Quit_x.value) < Formula.Name_len(Const.Name_Quit) and 0 < y_mouse  - Set.icon_Quit_y.value < Set.letter_height.value:
                    self.Name_icon =  Const.Name_Quit
                    return True
                if 0 < x_mouse - Set.author_point_x.value and 0 < y_mouse - Set.author_point_y.value:
                    self.Name_icon = Const.Name_Ezic
                    self.Ezic = True
                    return True
            elif event.type == pygame.QUIT:
                self.Name_icon =  Const.Name_Quit
                return True
        self.Name_icon = False
        return True
        
    def check_ezic(self):
        for event in pygame.event.get():
            if event.type == pygame.KEYDOWN and event.key == pygame.K_SPACE:
                self.Ezic = False
                return True
        
        
```

ENGINE

```python
class Engine:
    def __init__(self):
        self.Menu = Menu()
        self.Game = None
        
    def tick_menu(self):
        self.Menu.check_menu_event()
    
    def tick_game(self):
        self.Game.event_cycle()
        
    def tick_game_mouse(self):
        self.Game.mouse_update()
        
    def tick_return_game(self):
        self.Game.return_start_game()
```

CONTROLLER

```python
class Controller:
    def __init__(self):        
        self.pygame_init()

        self.UI = UI(self.screen)   
        self.Engine = Engine()
        
        self.main_running = True
        
    def pygame_init(self):
        pygame.init()
        self.screen = pygame.display.set_mode((Set.window_w.value, Set.window_h.value))
        self.Window = Window(Size(Set.window_w.value, Set.window_h.value), Const.Name_icon, Const.Icon_picture, self.screen)
        pygame.display.set_caption(self.Window.name)        
        icon = pygame.image.load(self.Window.icon)        
        pygame.display.set_icon(icon)


    def start_Game(self, Name_Game):
        if Name_Game == Const.Name_Gobblet:
            self.Engine.Game = Gobblet(self.Window)
        else:
            self.Engine.Game = Cross(self.Window)
        self.UI.Game = self.Engine.Game        
        self.Game_cycle()
        while self.Engine.Game.Data.return_game and self.main_running:
            self.Engine.tick_return_game()
            self.Game_cycle()
        
    def Game_cycle(self):
        self.Engine.Game.Data.Field.start_init()
        self.Engine.Game.Data.running = True
        while self.Engine.Game.Data.running:
            self.iteration()
        
    def iteration(self): 
        self.Engine.tick_game()
        
        self.UI.draw_click_counter()
        
        self.Engine.tick_game_mouse()
        
        if self.Engine.Game.Data.return_game:
            return True
        
        if not self.Engine.Game.Data.running:
            return True

        if self.Engine.Game.Data.game_over:
            self.UI.draw_game_over_update()
            return True
            
        self.UI.update_pygame_field()            
            
            
 #------------------------------           
    def Start(self):
        self.start_menu()
        
    def start_menu(self):
        while self.main_running:            
            self.Engine.tick_menu()
            Icon = self.Engine.Menu.Name_icon 
            if Icon:
                if Icon == Const.Name_Quit:
                    self.main_running = False
                elif Icon == Const.Name_Ezic:
                    self.update_Ezic()
                else:                
                    self.start_Game(Icon)
                    self.main_running = self.Engine.Game.Data.main_running
            else:
                self.UI.update_menu()
        self.exit()
        
    def exit(self):
        pygame.quit()

    def update_Ezic(self):
        while self.Engine.Menu.Ezic:
            self.Engine.Menu.check_ezic()
            self.UI.draw_Ezic()
```

MAIN

```python
def main():
    C = Controller()
    C.Start()
main()
```
